```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

df = pd.read_csv('expend.csv')
df
```

```python
categories = ["House", "Food", "Fun", "Education", "Travel"]
X = df[['Income', 'Savings'] + categories]
y = df['Expenses']

model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)


r2 = r2_score(y, y_pred)
print(f"R-squared (R2) score: {r2}")
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Expenses'])
plt.xlabel('Date')
plt.ylabel('Expenses')
plt.title('Expenses Over Time')
plt.grid(True)
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
result = seasonal_decompose(df['Expenses'], model='additive', period=365)
plt.figure(figsize=(12, 8))
plt.subplot(4, 1, 1)
plt.plot(df['Expenses'], label='Original Expenses')
plt.legend()

plt.subplot(4, 1, 2)
plt.plot(result.trend, label='Trend')
plt.legend()

plt.subplot(4, 1, 3)
plt.plot(result.seasonal, label='Seasonal')
plt.legend()

plt.subplot(4, 1, 4)
plt.plot(result.resid, label='Residual')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
from prophet import Prophet

prophet_df = df.reset_index()
prophet_df.rename(columns={'Date': 'ds', 'Expenses': 'y'}, inplace=True)
forecast_period = 30

model = Prophet()
model.fit(prophet_df)
future = model.make_future_dataframe(periods=forecast_period, freq='D')
forecast = model.predict(future)

fig = model.plot(forecast)
plt.xlabel('Date')
plt.ylabel('Expenses')
plt.title('Prophet Forecasting')
plt.show()

forecasted_values = forecast[['ds', 'yhat']].tail(forecast_period)
forecasted_values.rename(columns={'ds': 'Date', 'yhat': 'Forecasted_Expenses'}, inplace=True)
print(forecasted_values)
```
